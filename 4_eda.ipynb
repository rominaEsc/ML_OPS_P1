```python
import pandas as pd
import wordcloud
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv("data/movies_etl.csv", sep=",")
genres = pd.read_csv("data/genres.csv",sep =",")
movies_genres = pd.read_csv("data/movies_genres.csv",sep =",")
ml_df = pd.read_csv("data/ml_df.csv",sep =",")
sist_recomendacion_df = pd.read_csv("data/sist_recomendacion_df.csv",sep =",")
```

```python
text = sist_recomendacion_df.fillna(' ')
text['title'] = text['title'].astype('str')
text['overview']= text ['overview'].astype('str')
text['text'] = text['title']+ ' ' +text['overview']
stopwords = STOPWORDS
```

```python
text_title = ' '.join(text['title'])
text_overview = ' '.join(text['overview'])
```

```python
text.info()
```

```python
WC = WordCloud(
    background_color='white',
    stopwords=stopwords,
)
```

```python
for i in [text_title,text_overview]:
    WC.generate(i)
    plt.figure()
    plt.imshow(WC)
    plt.axis('off')
    plt.show()
```
