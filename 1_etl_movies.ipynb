# Duplicados

¿Hay filas duplicadas? ¿Cuántas?


```python
#filas duplicadas
(
    movies_raw
    .duplicated(keep=False)
    .sum()
)
```

```python
# Filas con campo id duplicado /  Una misma pelicula aparece varias veces en el dataframe.
(
    movies_raw
    .duplicated(subset='id',keep=False)
    .sum()
)
```

Eliminaremos las filas duplicadas y las que poseen ids duplicados. 
Ya que representan menos del 1% del dataset.
Pero se consultará para posteriores entregas.
